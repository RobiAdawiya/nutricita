```python
import pandas as pd

# Load dataset
df = pd.read_csv("nutrition.csv")

# Cek 5 baris pertama
print(df.head())

# Cek info kolom
print(df.info())

# Cek statistik deskriptif
print(df.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Contoh: Histogram kalori
plt.figure(figsize=(8, 4))
sns.histplot(df['kalori'], bins=20, kde=True)
plt.title("Distribusi Kalori")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Contoh: Histogram Protein
plt.figure(figsize=(8, 4))
sns.histplot(df['protein'], bins=20, kde=True)
plt.title("Distribusi Protein")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Contoh: Histogram kalori
plt.figure(figsize=(8, 4))
sns.histplot(df['lemak'], bins=20, kde=True)
plt.title("Distribusi Lemak")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Contoh: Histogram kalori
plt.figure(figsize=(8, 4))
sns.histplot(df['karbohidrat'], bins=20, kde=True)
plt.title("Distribusi Karbohidrat")
plt.show()
```

```python
# Heatmap korelasi
corr_matrix = df[['kalori', 'protein', 'lenak', 'karbohidrat']].corr()
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Korelasi Nutrisi")
plt.show()
```

```python
# Makanan dengan kalori tertinggi
print(df.nlargest(5, 'calories')[['name', 'calories']])

# Makanan dengan protein tertinggi
print(df.nlargest(5, 'proteins')[['name', 'proteins']])
```
